```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('./data/DiabetesData.csv')
columns = ['NPG', 'PGL', 'DIA', 'TSF', 'INS', 'BMI', 'DPF', 'AGE']
shuffled_data = dataset.sample(frac=1, random_state=36)
x = shuffled_data[columns]
y = shuffled_data['Diabetic']
```

```python

statistics = dataset.describe()

# Print the statistics table
print(statistics)
```

```python
import seaborn as sns

sns.countplot(x='Diabetic', data=dataset)
plt.xlabel('Diabetic Class')
plt.ylabel('Count')
plt.title('Distribution of Target Class')
plt.show()
```

```python
import sklearn as sk
from sklearn.model_selection import train_test_split

# split the data into 70% trainig nad 30% test  
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,  random_state = 1)
```

```python
from sklearn.tree import DecisionTreeRegressor

treeRegressor = DecisionTreeRegressor(random_state= 0)

treeRegressor.fit(x_train, y_train)
```

```python

prediction = treeRegressor.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score 

accuracy = accuracy_score(y_test, prediction)

print("Accuracy:", accuracy)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree


# Generate the decision tree plot
fig, ax = plt.subplots(figsize=(86, 82))
tree.plot_tree(treeRegressor, ax=ax)

# Show the plot
plt.show()
```
